# **Data Vizualization, Storytelling & Experimenting with charts**

```python
# Import Libraries

import matplotlib.pyplot as plt
import seaborn as sns
```

### Dataset Loading

```python
# Datasets Loading

daily_activity = pd.read_csv('Cleaned_Data/daily_activity_cleaned.csv', parse_dates=['ActivityDate'])
daily_calories = pd.read_csv('Cleaned_Data/daily_calories_cleaned.csv', parse_dates=['ActivityDay'])
sleep_day = pd.read_csv('Cleaned_Data/sleep_day_cleaned.csv', parse_dates=['SleepDate'])
weight_log = pd.read_csv('Cleaned_Data/weight_log_cleaned.csv', parse_dates=['WeightDate'])
hourly_steps = pd.read_csv('Cleaned_Data/hourly_steps_cleaned.csv', parse_dates=['ActivityDateTime'])
hourly_calories = pd.read_csv('Cleaned_Data/hourly_calories_cleaned.csv', parse_dates=['ActivityDateTime'])
```

#### Chart - 1

```python
# 1️ Activity Trends → Steps over time

sns.lineplot(x='ActivityDate', y='TotalSteps', data=daily_activity)
plt.title('Total Steps Over Time')
plt.xlabel('Date')
plt.ylabel('Steps')
plt.xticks(rotation=45)
plt.show()
```

##### What is/are the insight(s) found from the chart?

**Insight:**

* User activity levels vary over days.

* Some days show very low activity — device possibly not worn or inactive lifestyle days.
* No consistent upward or downward trend.

##### Will the gained insights help creating a positive business impact?

**Business Impact:**

* Opportunity to introduce daily step goals + motivational reminders.

* Could promote gamification → “Streaks” or “Challenges” to boost regular activity.

#### Chart - 2

```python
# 2 Bar Chart → Average Sleep by Weekday

sleep_day['Weekday'] = sleep_day['SleepDate'].dt.day_name()
sleep_by_weekday = sleep_day.groupby('Weekday')['TotalMinutesAsleep'].mean().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])

sns.barplot(x=sleep_by_weekday.index, y=sleep_by_weekday.values)
plt.title('Average Sleep Duration by Weekday')
plt.ylabel('Minutes Asleep')
plt.xticks(rotation=45)
plt.show()
```

##### What is/are the insight(s) found from the chart?

**Insight:**

* Users tend to sleep slightly more on weekends (Sat/Sun).

* Sleep duration drops during weekdays → possible work stress or routine factors.

##### Will the gained insights help creating a positive business impact?

**Business Impact:**

* Strava can promote sleep coaching content during weekdays.

* Can provide personalized weekday sleep tips → improve sleep quality on workdays.



#### Chart - 3

```python
# 3 Steps vs Calories burned → Scatter plot

sns.scatterplot(x='TotalSteps', y='Calories', data=daily_activity)
plt.title('Steps vs Calories Burned')
plt.xlabel('Steps')
plt.ylabel('Calories')
plt.show()
```

##### What is/are the insight(s) found from the chart?

**Insight:**

* Strong positive correlation between steps taken and calories burned → confirms expected fitness behavior.

* Outliers with low steps but high calories → likely due to other activities (cycling, strength training, etc.).

##### Will the gained insights help creating a positive business impact?

**Business Impact:**

* Strava can emphasize to users that steps alone do not tell full calorie story.

* App can promote tracking other forms of exercise to improve accuracy of calorie estimation.

#### Chart - 4

```python
# 4 Horizontal Bar Chart → Top 5 Users by Average Daily Steps

top_users = daily_activity.groupby('Id')['TotalSteps'].mean().sort_values(ascending=True).tail(5)

top_users.plot(kind='barh', color='skyblue')
plt.title('Top 5 Users by Average Daily Steps')
plt.xlabel('Average Steps')
plt.ylabel('User ID')
plt.show()
```

##### What is/are the insight(s) found from the chart?

**Insight:**

* Certain users (top 5) consistently outperform others in step count.

* These users demonstrate high engagement with activity tracking.

##### Will the gained insights help creating a positive business impact?

**Business Impact:**

* Strava can target top users with premium subscriptions, early access to new features, or reward programs.

* These users can be ambassadors / influencers → promote Bellabeat to peers.

#### Chart - 5

```python
# 5 Sleep Patterns → Sleep duration over time

sns.lineplot(x='SleepDate', y='TotalMinutesAsleep', data=sleep_day)
plt.title('Sleep Duration Over Time')
plt.xlabel('Date')
plt.ylabel('Minutes Asleep')
plt.xticks(rotation=45)
plt.show()
```

##### What is/are the insight(s) found from the chart?

**Insight:**

* Sleep duration fluctuates over time → not always consistent.

* Some users show periods of reduced sleep → opportunity for intervention.

##### Will the gained insights help creating a positive business impact?

**Business Impact:**

* Strava can offer personalized sleep trends reports.

* App can trigger alerts when user's sleep drops below healthy thresholds → proactive wellness support.

#### Chart - 6

```python
# 6 Pie Chart → Active Minutes Breakdown

active_minutes_avg = daily_activity[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']].mean()

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
active_minutes_avg.plot(kind='pie', autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Average Distribution of Daily Activity Minutes')
plt.ylabel('')
plt.show()
```

##### What is/are the insight(s) found from the chart?

**Insight:**

* Sedentary time dominates daily minutes (~70-80%) → similar to known population trends.

* VeryActive and FairlyActive minutes form a smaller portion of the day.

##### Will the gained insights help creating a positive business impact?

**Business Impact:**

* Strava can promote Active Break reminders → combat sedentary behavior.

* App can introduce “Active Minutes” challenges to gradually shift users towards more active lifestyle.

#### Chart - 7

```python
# 7 Weight Trends → Line plot

plt.figure()
sns.lineplot(x='WeightDate', y='WeightKg', data=weight_log)
plt.title('Weight Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Weight (Kg)')
plt.xticks(rotation=45)
plt.show()
```

##### What is/are the insight(s) found from the chart?

**Insight:**

* Users tracking weight show mostly stable trends → no drastic changes.

* Some variation suggests ongoing weight management efforts.

##### Will the gained insights help creating a positive business impact?

**Business Impact:**

* Strava can promote long-term weight tracking → celebrate small positive changes.

* App can encourage regular weight logging → helps users stay engaged and mindful.
